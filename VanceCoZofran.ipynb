## Process the allVomit dataframe so that there are no rows with the same 'DISPID', and such that each 'DISPID' instead has a set for the possible 'NAME' and 'ALLERGY NAME' values:

```python
allVomit2 = allVomit.groupby(['DISPID', 'SEX', 'RACE']).agg({
    'NAME': lambda x: list(set(x)),  # Use set to avoid duplicates, then convert to list
    'ALLERGY NAME': lambda x: list(set(x))  # Use set to avoid duplicates, then convert to list
}).reset_index()
allMeds
```

## This counts antiemetic administrations by race. It's O(n^2) and might take a few minutes to run

```python
antiEmeticB = 0 #Number of Nausea/Vomiting DISPID's black that get antiEmetic
antiEmeticN = 0 #Number of Nausea/Vomiting DISPID's nonblack that get antiEmetic

for index, row in allVomit2.iterrows():
    print(index)
    print(f"BlackantiEmetic: {antiEmeticB} WhiteantiEmetic:{antiEmeticN}")
    for indexMed, rowMed in allMeds.iterrows():
        if row['DISPID'] == rowMed['DISPID']:
            if rowMed['MEDICATION'] in ("Ondansetron","Ondansetron Oral Disintegrating Tablet (ODT)","Promethazine","Procholorperazine","Metoclopramide"):
                if row['RACE'] == "Black, Non-Hispanic":
                    antiEmeticB += 1
                    break
                elif row['RACE'] == "White, Non-Hispanic":
                    antiEmeticN += 1
                    break
```

## This counts only Ondansetron administrations by race. It's O(n^2) and might take a few minutes to run

```python
zofranB = 0 #Number of Nausea/Vomiting DISPID's black that get Zofran
zofranN = 0 #Number of Nausea/Vomiting DISPID's nonblack that get Zofran

for index, row in allVomit2.iterrows():
    print(index)
    print(f"BlackantiEmetic: {zofranB} WhiteantiEmetic:{zofranN}")
    for indexMed, rowMed in allMeds.iterrows():
        if row['DISPID'] == rowMed['DISPID']:
            if rowMed['MEDICATION'] in ("Ondansetron","Ondansetron Oral Disintegrating Tablet (ODT)"):
                if row['RACE'] == "Black, Non-Hispanic":
                    zofranB += 1
                    break
                elif row['RACE'] == "White, Non-Hispanic":
                    zofranN += 1
                    break
```

## Number of black and white patients:

```python
#Number of Black, Non-Hispanic and White, Non-Hispanic Individuals in data:
numBlack = 0
numWhite = 0
for race, count in allVomit2['RACE'].value_counts(dropna=False).items():
    if race == "Black, Non-Hispanic":
        numBlack = count
    elif race == "White, Non-Hispanic":
        numWhite = count

print(f"Number Black patients: {numBlack}, Number White: {numWhite}")
```

## Proportion and two-proportion Z-hypothesis test for race and ALL ANTIEMETIC administration:

```python
print(f"Proportion Black who received antiemetics: {antiEmeticB/numBlack*100}%")
print(f"Proportion White who received antiemetics: {antiEmeticN/numWhite*100}%")
#two-proportion z-test
count = np.array([antiEmeticB, antiEmeticN])
nobs = np.array([numBlack, numWhite])

stat, pval = sm.stats.proportions_ztest(count, nobs)

print(f"Z-statistic: {stat}")
print(f"P-value: {pval}\n")


alpha = 0.05  #Significance level
if pval < alpha:
    print("The difference in proportions is thus statistically significant.")
else:
    print("The difference in proportions is thus not statistically significant.")
```

## Proportion and two-proportion Z-hypothesis test for race and ONDANSETRON administration:

```python
print(f"Proportion Black who received zofran: {zofranB/numBlack*100}%")
print(f"Proportion White who received zofran: {zofranN/numWhite*100}%")
#two-proportion z-test
count = np.array([zofranB, zofranN])
nobs = np.array([numBlack, numWhite])

stat, pval = sm.stats.proportions_ztest(count, nobs)

print(f"Z-statistic: {stat}")
print(f"P-value: {pval}\n")


alpha = 0.05  #Significance level
if pval < alpha:
    print("The difference in proportions is thus statistically significant.")
else:
    print("The difference in proportions is thus not statistically significant.")
```
